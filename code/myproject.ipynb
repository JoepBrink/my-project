```python
import pandas as pd

import numpy as np

from matplotlib import pyplot as plt

global pd

global np

data = pd.read_csv('../data/mydata.csv')

global data
```

```python
for i in range(0, 8):

    a = data.a[i]

    xcen = data.perihelion[i] - a

    ycen = 0

    b = data.b[i]

    t = np.linspace(0, 2*np.pi, 100)

    plt.plot(xcen + a * np.cos(t), ycen + b * np.sin(t))

    plt.grid(color = "black", linestyle = "--")

plt.savefig("../results/plot.png")
```
